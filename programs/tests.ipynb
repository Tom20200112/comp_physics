```python
def factorial(n: int):
    n_factorial = 1
    while n:
        n_factorial = n_factorial * n
        n = n -1
    return n_factorial

print(factorial(23)-factorial(21))
```

```python
import numpy as np
def proper_divisors(x: int):
    l=[]
    for i in range(1,x):
        if x%i==0:
            l.append(i)

    return l
    
    
n=1
while n:
    if np.sum(proper_divisors(n))==3*n:
        break
    n+=1
    
print(n)
```

```python
from itertools import combinations 
from math import sqrt
from typing import Callable
import numpy as np
import morse

def pairwise_potential_with_args(potential: Callable[[float], float], xss: [[float]], *potential_args) -> float:
    return sum(
        potential(
            sum((xia - xja)**2 for xia, xja in zip(xi, xj))**0.5, *potential_args
        ) for xi, xj in combinations(xss, 2)
    )
potential_args = (6,1, 1.0)

morse13 = np.loadtxt("morse_n_13_rho_6.txt")

pairwise_potential_with_args(morse.morse_potential,morse13, *potential_args)
```

for $f(x)=exp(x)-1$, 
$$
\kappa(f,x)=\left\lvert\frac{x f'\left(x\right)} {f\left(x\right)}\right\rvert=\left\lvert \frac{xexp(x)}{exp(x)-1}\right\rvert
$$
when $x\rightarrow0$, $\frac{x}{exp(x)-1}\rightarrow1$ and $exp(x)\rightarrow1$, therefore $\kappa(f,x)\rightarrow1$

```python
x=np.logspace(-30,0,200)
fig,ax=plt.subplots(1,1)
ax.plot(x,np.expm1(x),label="expm1")
ax.plot(x,np.exp(x)-1,label="exp-1")
ax.legend(loc="upper left")

ax.set_xscale('log')
ax.set_yscale('log')
plt.xlabel("x")
plt.ylabel("exmp1(x) and exp(x)-1")
plt.savefig("expm1.png")
```

```python
def trapezium_int(t: [float], v: [float])-> float:
    if len(t)!=len(v):
        raise ValueError("The lenth of t and v should be equal to each other")
    
    int=0
    for i in range(0,len(t)-1):
        int+=(t[i+1]-t[i])*(v[i+1]+v[i])/2
    return int
```

```python
def simpson_int(t:[float],v:[float])->float:
    if len(t)%2==0 or len(t)!=len(v):
        raise ValueError("The lenth of t and v should be odd and equal to each other")
    int =0
    for i in range(0,(len(t)-1)//2):
        h=t[2*i+1]-t[2*i]
        int+=h/3*(v[2*i]+4*v[2*i+1]+v[2*i+2])
        
    return int
```

```python
t=[0,68,136,204,272,340,408,476,544]
v=[0,404,867,1407,2059,2877,3980,5677,9520]

trapezium_int(t,v)
```
